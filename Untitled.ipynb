```python
def DBconnect():#Connection to database
    conn = sqlite3.connect("itdaa_db1.db")
    return conn
 
def SQLTable(conn):#Creating table
    cur = conn.cursor()
    cur.execute("DROP TABLE IF EXISTS heart")
    data = """
    CREATE TABLE heart(    
        age           integer,
        sex           text,
        cp            integer,
        trestbps      integer,
        chol          integer,
        fbs           integer,
        restecg       integer,
        thalach       integer,
        exang         integer,
        oldpeak       real,
        slope         integer,
        ca            integer,
        thal          integer,
        target        text
        )"""

    cur.execute(data)
    print("Database has been created")
    conn.commit()
    
 #convert to actual value shown on Dataset Overview instead of showing digit
def convertNominal(df):
    df['sex'] = df['sex'].astype(str)
    df.loc[df['sex'] == '1', 'sex'] = 'male'
    df.loc[df['sex'] == '0', 'sex'] = 'female'
    
    df['cp'] = df['cp'].astype(str)
    df.loc[df['cp'] == '0', 'cp'] = 'typical angina'
    df.loc[df['cp'] == '1', 'cp'] = 'atypical angina'
    df.loc[df['cp'] == '2', 'cp'] = 'non-angina pain'
    df.loc[df['cp'] == '3', 'cp'] = 'asymptomatic'
    
    df['target'] = df['target'].astype(str)
    df.loc[df['target'] == '1', 'target'] = 'heart disease'
    df.loc[df['target'] == '0', 'target'] = 'healthy'
    
    df['fbs'] = df['fbs'].astype(str)
    df.loc[df['fbs'] == '1', 'fbs'] = 'true'
    df.loc[df['fbs'] == '0', 'fbs'] = 'false'
    
    #restecg missing
    df['restecg'] = df['restecg'].astype(str)
    df.loc[df['restecg'] == '0', 'restecg'] = 'normal'
    df.loc[df['restecg'] == '1', 'restecg'] = 'abnormal'
    df.loc[df['restecg'] == '2', 'restecg'] = 'ventricular hypertrophy'
    
    df['exang'] = df['exang'].astype(str)
    df.loc[df['exang'] == '1', 'exang'] = 'yes'
    df.loc[df['exang'] == '0', 'exang'] = 'no'
    
    df['slope'] = df['slope'].astype(str)
    df.loc[df['slope'] == '0', 'slope'] = 'upsloping'
    df.loc[df['slope'] == '1', 'slope'] = 'flat'
    df.loc[df['slope'] == '2', 'slope'] = 'downsloping'
    
    #ca missing
    
    df['thal'] = df['thal'].astype(str)
    df.loc[df['thal'] == '0', 'thal'] = 'unknown'
    df.loc[df['thal'] == '1', 'thal'] = 'normal'
    df.loc[df['thal'] == '2', 'thal'] = 'fixed defect'
    df.loc[df['thal'] == '3', 'thal'] = 'reversible defect'
    return df

def CSVImport(conn):
    cur = conn.cursor()
    df = pd.read_csv("C:/Users/Emmanuel/Documents/Data Science/ITDAA4-12 Data Mining and Data Administration/Project/heart.csv",header = 0,sep=";")
    df=convertNominal(df)
   
    for index in range(len(df)):
        cur.execute("INSERT INTO heart(age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal,target) VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?)",(df['age'].loc[index],df['sex'].loc[index],df['cp'].loc[index],df['trestbps'].loc[index],df['chol'].loc[index],df['fbs'].loc[index], df['restecg'].loc[index],df['thalach'].loc[index],df['exang'].loc[index],df['oldpeak'].loc[index],df['slope'].loc[index],df['ca'].loc[index],df['thal'].loc[index],df['target'].loc[index]))
        conn.commit()
    print('\n Records added')
   
def DB_Read(conn):
    cur = conn.cursor()
    #cur.execute("SELECT * FROM  students;") #select table names
    #data = cur.fetchall()
    data = pd.read_sql("SELECT * FROM  heart;",conn)
    conn.commit()
    return data
    
conn = DBconnect()
#print(conn)
SQLTable(conn)
CSVImport(conn)

#check for nulls
#print(pd.read_sql("SELECT COUNT(*) FROM heart WHERE age+ sex + cp+ trestbps+ chol+ fbs +restecg +thalach+  exang+oldpeak+slope +ca +thal+target  IS NULL;",conn))
#conn.commit()


#store table data in dataframe
df_table=DB_Read(conn)
print(df_table)
```

```python
## Association Rule Mining  Apriori
#start#
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules


#change fbs for algorithm
mapping_fbs = {
    'true': 'fbs>120 mg/dl',
    'false': 'fbs<120 mg/dl'
}

#change exang for algorithm
mapping_exang = {
    'yes': 'exang yes',
    'no': 'exang no'
}

df_AsRule=DB_Read(conn)
df_AsRule['fbs'] = df_AsRule['fbs'].replace(mapping_fbs)
df_AsRule['exang'] = df_AsRule['exang'].replace(mapping_exang)

nominal_columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal', 'target']
df_selected = df_AsRule[nominal_columns]
print(df_selected.head())
```

```python
# Combine items into transactions, handling missing values
transactions = df_selected.apply(lambda row: row.dropna().tolist(), axis=1).tolist()
# Convert all items in transactions to strings
transactions = [[str(item) for item in transaction] for transaction in transactions]
# Display the transactions
print(transactions)
```

```python
# Initialize the transaction encoder
t_encoder = TransactionEncoder()

# Fit and transform the transactions
t_encoder_array = t_encoder.fit(transactions).transform(transactions)

# Convert the array to a DataFrame
df_encoded = pd.DataFrame(t_encoder_array, columns=t_encoder.columns_)

# Display the encoded DataFrame
print(df_encoded)
```

```python
# Apply the Apriori algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Display the rules
print(rules)
```

```python
#save model

import pickle 
  
# Save the trained model as a pickle string. 
# Save frequent itemsets and rules to a pickle file
with open('as_rule_apriori_model.pkl', 'wb') as f:
    pickle.dump((frequent_itemsets, rules), f)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

df_km = pd.DataFrame(DB_Read(conn))
# prep the data
prep = ColumnTransformer(
    transformers=[
        ('sex', OneHotEncoder(), ['sex']),
        ('cp', OneHotEncoder(), ['cp']),
        ('fbs', OneHotEncoder(), ['fbs']),
        ('restecg', OneHotEncoder(), ['restecg']),
        ('exang', OneHotEncoder(), ['exang']),
        ('slope', OneHotEncoder(), ['slope']),
        ('thal', OneHotEncoder(), ['thal']),
        ('ca', OneHotEncoder(), ['ca']),
        ('target', OneHotEncoder(), ['target'])
    ],
    remainder='passthrough'
)
X = prep.fit_transform(df_km)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(df_km[['thalach']])
df_km['thalach'] = scaler.transform(df_km[['thalach']])

scaler.fit(df_km[['age']])
df_km['age'] = scaler.transform(df_km[['age']])
```

```python
 #Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 50):
    kmeans2 = KMeans(n_clusters = i)
    kmeans2.fit(X)
    # inertia method returns wcss for that model
    wcss.append(kmeans2.inertia_)
    
#[1,2,3,...11]
plt.figure(figsize=(10,5))
sns.lineplot(x= range(1, 50), y= wcss,marker='o',color='red')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=3)
df_km['cluster'] = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_
#print(centroids)
df_km.head()
```

```python
df1 = df_km[df_km.cluster==0]
df2 = df_km[df_km.cluster==1]
df3 = df_km[df_km.cluster==2]

plt.scatter(df1['age'],df1['thalach'],color='green')
plt.scatter(df2['age'],df2['thalach'],color='red')
plt.scatter(df3['age'],df3['thalach'],color='black')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('age')
plt.ylabel('thalach')
plt.legend()
```
